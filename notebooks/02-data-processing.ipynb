```python
import pandas as pd
import os
```

```python
df = pd.read_csv('../data/final/combined.csv', index_col=False)
df.head()
```

```python
features = ['W-L%', 'FG%', '3P%', 'FT%', 'FTr', '3PAr', 'TS%', 'TRB%', 'AST%', 'STL%', 'BLK%', 'eFG%', 'TOV%', 'ORB%', 'FT/FGA']
features2 = ['TP', 'OP','FG', 'FGA', '3P', '3PA', 'FT', 'FTA', 'ORB', 'TRB','DRB', 'AST', 'STL', 'BLK', 'TOV', 'PF']
for i in range(2014,2024):
    df = pd.read_csv(f'../data/final/{i}-school-stats.csv', index_col=False)
    for feature in features2:
        temp = df[feature].items()
        for index, item in temp:
            if item > 1000:
                df[feature][index] = item / 1000
            elif item > 100:
                df[feature][index] = item / 100

        df[feature] = df[feature].astype(float)
    
    df.to_csv(f'../data/newfinal/{i}-school-stats.csv', index=False)
```

```python
path = '../data/newfinal/'
newPath = '../data/newfinal/'

big = pd.DataFrame()
for file in os.listdir(path):
    if file.endswith('.csv') and file != 'combined.csv':
        df = pd.read_csv(path+file)
        df.fillna('NA', inplace=True)
        df['YEAR'] = file[:4]
        big = pd.concat([df, big], ignore_index=True)
        df.to_csv(newPath+file, index=False)

big.to_csv(f'{newPath}combined.csv', index=False)
```

```python
df = pd.read_csv('../data/newfinal/combinedfix.csv')


# Create a mask for columns that need to be fixed
mask = ['TRB']
# Iterate over the selected columns and apply the fix to each value that needs it

temp = df.loc[df['YEAR'].isin([2022,2023])]

for col in mask:
    temp[col] = temp[col].div(temp['G'])

df.update(temp)

df.to_csv('../data/newfinal/combinedfix.csv', index=False)
```
